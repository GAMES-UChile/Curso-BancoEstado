### 0.1 Historia 
Python fue concevido en los 1980s por Guido van Rossum en los países bajos como sucesor de el lenguaje ABC. Sin embargo, su implementación comenzó en 1989. Van Rossum continuó como el desarrollador principal de Python hasta el 12 de julio de 2018. Desde enero 2019, Python es dirigido por un comité de 5 miembros (que incluyen a van Rossum).

Python 2.0 fue lanzado el 16 de octubre de 2000 y Python 3.0 el 3 de deciembre de 2008, esta última fue una revisión mayor que ya no es compatible hacia versiones anteriores. Sin embargo, muchas de sus funciones pueden ser cargadas desde versiones anteriores (cuando vean `import from future`, es precisamente esto lo que están haciendo. Ni siquiera la sintaxis es completamente compatible entre Python 2 y 3, por lo que se ha creado la utilidad 2to3, la cual traduce código Python 2 a Python 3. 

El fin del soporte de Python 2.7 fue inicialmente declarado para 2015 pero postergado para 2020 debido a la gran cantidad de código existente que no podía fácilmente migrarse a Python 3. Luego de 2020, Python 2 no tendrá más soporte. 

### 0.2 Filosofía

Python es un lenguaje muy completo que incluye una cantidad enorme de recursos, siendo de utilidad en amplio dominio de aplicación. Esto es posible porque, en vez de incluir toda esta funcionalidad en su core, Python es altamente extendible. Esta capacidad lo hace muy popular para conectar con otros recursos. La visión de van Rossum era precisamente hacer un lenguaje simple pero que puede _conversar_ con otros.

Python da prioridad a una sintaxis clara, simple y práctica, donde los programadores pueden elegir su propia metodología. De acuerdo a Python, "debe haber una---y preferencemente solo una---forma obvio de hacerlo". En general, en Python se va a dar prioridad a claridad y no velocidad. Sin embargo, si la velocidad es importante, existen extensiones que permiten usar código C compilado desde Python.

Un objetivo importante de Python que su uso sea entrentenido para los desarrolladores y usuarios, lo cual es reflejado por su nombre, inspirado por la comedia británica _Monty Python), y por otros ejemplos tutoriales como _spam and eggs_ (del sketch de Monty Python) en vez del clásico _foo and bar_.

Un neologismo en la comunidad Python es **pytónico**, o _pythonic,_ el que puede tener un amplia rango de significados relacionados al estilo de programación. El decir que algo es **pytónico** es decir que usa Python de acuerdo a su filosofía, que es natural y fluido en la programación y sigue la filosofía minimalista con énfasis en legibilidad. En contraste, código que es difícil de entender o bien que parece ser una transcripción cruda desde otro lenguaje, será llamado **anti-pytónico**, o _unpythonic_.

Para acceder a la filosofía de Python, podemos importar el Zen de Python mediante `import this`

### 2.B Cargar datos

Numpy permite cargar archivos en formato .txt, .csv

**Nota**: También se puede usar [Pandas](https://pandas.pydata.org/), una librería especializada en el manejo de archivos csv, excel y HDF5 entre otros para bases de datos más completas y no necesariamente compuestas de números.

> Antes de cargar nuestros datos, creemos una base de datos y guardémosla, luego la cargaremos.
 Para crear nuestros datos usemos numpy arrays. 

````python
# media
mu = np.array([3, 1])
# matriz covarianza
cov = np.array([[1, 0.5], [0.5, 1]])
# número de muestras
n = 50

samples = np.random.multivariate_normal(mean=mu, cov=cov, size=(50))
print('Las muestras totales son de tamaño', samples.shape)
````

> Ahora que tenemos nuestros datos podemos guardarlos.

````python
nombre_archivo = 'datos.csv'
np.savetxt(nombre_archivo, X=samples, delimiter=',')
````

> Ahora que tenemos nuestros datos, podemos cargarlos y empezar a trabajar con ellos.


> __Primero__: dividamos las muestras según eje, donde llamaremos las primera dimensión  $X$ y la segunda $Y$. Luego utilizemos numpy para obtener intuición inicial de los datos.

````python
datos = np.loadtxt('datos.csv', delimiter=',')
print('Datos cargados tienen tamaño', datos.shape)
````

````python
# asignar muestras en X
x = datos[:, 0]
# asignar muestas del eje Y
y = datos[:, 1]

print('El arreglo en el eje x tiene un tamño', x.shape)
````

> Obtengamos la media de cada eje y veamos que tan cerca está del valor real.

````python
# obtener media
media_x = x.mean()
print('Media empírica en X', media_x)

# media en y la calculamos de forma directa en el print
print('Media empírica en Y', y.mean())
````

````python
# arreglo equespaciado de -5 a 5 con 200 puntos
t = np.linspace(-5, 5, num=200)
# frecuencia
f = 0.5
signal = np.sin(t * 2* np.pi * f)

# graficamos
plt.plot(t, signal)
# nombre del eje X
plt.xlabel('Tiempo')
# nombre eje Y
plt.ylabel('Valor')
# mostrar figura
plt.show()
````

### 4.B Optimización

> Utilizemos scipy para optimizar una función, para esto primero crearemos una función de varios parametros y luego minimizemos respecto a un subconjunto de estos.

> Definamos nuestra función a optimizar, en este caso será de la forma 
$$ J = \alpha \left[(x - \beta)^2 + (y - \gamma)^2 \right]$$
Donde $\alpha, \beta, \gamma$ son parametros __fijos__ del a función, donde minimizaremos respecto a $x$ e $y$.
Para distinguir entre los parametros fijos de nuestra función, el primer argumento será un ndarray con $z=(x,y)$, y los consecuentes argumentos serán los demás parámetros.
De esta forma sabemos que el mínimo está en $(x, y) = (\alpha, \beta)$ con $J=0$.

````python
def funcion_obj(z, alpha, beta, gamma):
    """
    Función objetivo a optimizar.
    """
    x = z[0]
    y = z[1]
    
    J = (x - beta)**2 + (y - gamma)**2
    J = J * alpha
    return J

# probemos la funcion
a = 0.5 # alpha
b = 20  # beta
c = -10 # gamma

x = 543
y = 500
z = np.array([x, y])
print('Valor función objetivo', funcion_obj(z, a, b, c))
````

````python
# importemos el optimizador
from scipy.optimize import minimize
````

> **Ahora optimizemos**

````python
# punto inicial al optimizar
z_inicial = np.array([500, 500])

# opciones extras
opciones = {'maxiter':1000, 'disp':True}

res = minimize(
        fun=funcion_obj,          # funcion objetivo
        x0=z_inicial,             # punto inicial
        args=(a, b, c),           # argumentos fijos
        method='L-BFGS-B',        # metodo optimización
        options=opciones          # argumentos extras
        )
````

> Ahora que tenemos nuestro resultado podemos obtener que valor obtuvo el optimizador

````python
# valor de z=(x,y) optimo
z_optimo = res.x
print('Valor de (x,y) optimo encontrado:', z_optimo)

# valor de la funcion objetivo en el optimo
J_optimo = res.fun
print('Valor de funcion objetivo minima encontrada:', J_optimo)
````

> Por ultimo, intentemos graficar esta función con un mapa de colores, para esto hagamos una malla de valores para X e Y y luego grafiquemos.

````python
# crear equiespaciado para evaluar función
x = np.linspace(-50, 50, 100)
y = np.linspace(-50, 50, 100)

# crear malla
xx, yy = np.meshgrid(x, y)
# evaluar malla
jj = funcion_obj([xx, yy], a, b, c)
````

> Ahora grafiquemos la función, otra cosa importante que haremos es guardar la figura en graficos vectorizados para evitar perder resolución.

````python
plt.figure(figsize=(8, 6))
# graficar contorno
plt.contour(xx, yy, jj, 120, cmap='RdGy')
plt.plot(z_optimo[0], z_optimo[1], marker='*', markersize=20, color='g', label='Óptimo encontrado')
# muestra legenda
plt.legend()

# fija nombre de ejes
plt.xlabel('X')
plt.ylabel('Y')

# muestra barra de color
plt.colorbar()

# elimina espacios inncesarios en figura
plt.tight_layout()

plt.savefig('figura.pdf')
````
