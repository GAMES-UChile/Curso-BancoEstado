### 2.1 Optimización
https://docs.scipy.org/doc/scipy/reference/tutorial/optimize.html
El paquete `scipy.optimize` provee algoritmos comunmente usados en optimización, entre ellos:

- Rutinas de fuerza bruta
- Mínimos cuadrados y algoritmos de ajuste de curvas

En el ejemplo que se muestra a continuación se minimiza la función de Rosenbrok de $N$ variables dada por
\begin{equation}
    \sum_{i=1}^{N-1} 100*(x_{i+1} - x^2_i)^2 + (1-x_i)^2.
\end{equation}
El algoritmo usado puede ser 'nelder-mead' o bien 'powell'. Lo bueno de estos métodos es que sólo se necesita evaluar la función y no usa evaluación del gradiente. (es por ello que es más lenta).

```python
from scipy.optimize import minimize

def rosen(x):
    return sum(100.0 * (x[1:] - x[:-1]**2.0)**2.0 + (1 - x[:-1])**2.0)
```

```python
%matplotlib notebook
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

xs, ys = np.meshgrid(np.linspace(-1,1,50), np.linspace(-1,1,50))
zs = np.empty((xs.shape[0], xs.shape[1]))
for j in range(xs.shape[0]):
    for i in range(xs.shape[1]):
        zs[j,i] = rosen(np.array([xs[j,i], ys[j,i]]))

fig = plt.figure(figsize=(8,5))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(xs, ys, zs, alpha=0.8, cmap=plt.cm.viridis, vmin=-50, vmax=300)
ax.set_title('Función de costo Rosenbrock')
ax.set_xlabel('Parámetro 1')
ax.set_ylabel('Parámetro 2')
ax.set_zlabel('Costo')

plt.draw()
```

```python
%%time
x0 = np.array([1.3, 0.7, 0.8, 1.9, 1.2])
res = minimize(rosen, x0, method='BFGS', options={'maxiter': 100, 'disp': True})
print(res)
```

```python
%%time
res = minimize(rosen, x0, method='powell', options={'xtol': 1e-8, 'disp': True})
print(res)
```

```python
%%time 
res = minimize(rosen, x0, method='nelder-mead', options={'xtol': 1e-8, 'disp': True})
print(res)
```

Para obtener un mejor desempeño (converger en menos iteraciones al óptimo), se implementará un método para optimizar que usa el gradiente de la función, por ello se define la función rosen_der, que corresponde al gradiente.

Luego se llama al algoritmo 'BFGS', el cual recibe como argumento el gradiente de la función en 'jac'.

En caso de no recibir explicitamente el jacobiano, el método lo obtiene numéricamente por diferencias finitas.

Para minimizaciones de problemas más grandes, se puede usar el método de Newton-CG, el cual necesita el Hessiano. No se explicita aquí.

```python
def rosen_der(x):
    xm = x[1:-1]
    xm_m1 = x[:-2]
    xm_p1 = x[2:]
    der = np.zeros_like(x)
    der[1:-1] = 200*(xm-xm_m1**2) - 400*(xm_p1 - xm**2)*xm - 2*(1-xm)
    der[0] = -400*x[0]*(x[1]-x[0]**2) - 2*(1-x[0])
    der[-1] = 200*(x[-1]-x[-2]**2)
    return der
```

```python
%%time 
res = minimize(rosen, x0, method='CG', jac=rosen_der, options={'disp': True})
print(res)
```

```python
%%time 
res = minimize(rosen, x0, method='BFGS', jac=rosen_der, options={'disp': True})
print(res)
```

```python
%%time 
res = minimize(rosen, x0, method='L-BFGS-B', jac=rosen_der, options={'disp': True})
print(res)
```

```python
%%time 
res = minimize(rosen, x0, method='Newton-CG', jac=rosen_der, options={'disp': True})
print(res)
```

#### Datos de aerolínea

```python
%matplotlib inline 
import math
import matplotlib.pyplot as plt
from scipy.optimize import fmin_l_bfgs_b as fmin
plt.rcParams['figure.figsize'] = (15,5)

# cargar los datos
data = np.loadtxt('passenger_data.txt')
x = data[:,0]
y = data[:,1]

train_ratio = 0.8  # porcentaje para entrenamiento versus test
idx_test = int(len(y)*train_ratio)

x_train = x[:idx_test]
y_train = y[:idx_test]
x_test = x[idx_test:]
y_test = y[idx_test:]


# parte polinomial
x_tilde = np.vstack((x_train, x_train**2, x_train**3, np.ones(len(x_train))))
rho = .001
V = np.linalg.inv(x_tilde@x_tilde.T + rho*np.identity(4))
W = np.sum(x_tilde.T*y_train[:,np.newaxis], axis=0) @ V


# parte seno
def nll_seno(theta, W, y_obs, x_obs):
    (alfa1, lambda1, w1, sigma), phi1 = np.exp(theta[:4]), theta[4]

    n_obs = len(y_obs)
    nll = n_obs*0.5*math.log(2*np.pi*sigma**2)
    for i in np.arange(n_obs):
        y_t = np.dot([x_obs[i], x_obs[i]**2, x_obs[i]**3, 1], W)
        y_t += alfa1*np.exp(x_obs[i]*lambda1)*np.sin(w1*x_obs[i]+phi1)
        nll += (y_obs[i] - y_t)**2/(2.0*sigma**2)
    return nll

theta_seno = np.log([10, 0.01, 0.5, 1, 1.1])  # alfa1, lambda1, w1, sigma, phi1
theta_seno, _, _ = fmin(nll_seno, theta_seno, None, (W, y_train, x_train),
                        approx_grad=True, disp=1, factr=4500, maxiter=100000)


# parte mixto
def nll_mixto(theta, theta_seno, W, y_obs, x_obs):
    (alfa1, lambda1, w1), phi1 = np.exp(theta_seno[:3]), theta_seno[4]
    (alfa2, lambda2, w2, sigma), phi2 = np.exp(theta[:4]), theta[4]
    
    n_obs = len(y_obs)
    nll = n_obs*0.5*np.log(2*np.pi*sigma**2)
    for i in np.arange(n_obs):
        y_t = np.dot([x_obs[i], x_obs[i]**2, x_obs[i]**3, 1], W)
        y_t += alfa1*np.exp(x_obs[i]*lambda1)*np.sin(w1*x_obs[i]+phi1)
        y_t += alfa2*np.exp(x_obs[i]*lambda2)*np.sin(w2*x_obs[i]+phi2)
        nll += (y_obs[i] - y_t)**2/(2.0*sigma**2)
    return nll

theta_mixto = np.log([10, 0.1, 1, 1, 1])  # alfa2, lambda2, w2, sigma, phi2
theta_mixto, _, _ = fmin(nll_mixto, theta_mixto, None, (theta_seno, W, y_train, x_train),
                         approx_grad=True, disp=1, factr=4500, maxiter=100000)

                         
# construir y graficar la predicción
(alfa1, lambda1, w1), phi1 = np.exp(theta_seno[:3]), theta_seno[4]
(alfa2, lambda2, w2, sigma), phi2 = np.exp(theta_mixto[:4]), theta_mixto[4]

y_pred = np.zeros(len(y))
for i in range(len(y_pred)):
    y_pred[i] = np.dot([x[i], x[i]**2, x[i]**3, 1], W)
    y_pred[i] += alfa1*np.exp(x[i]*lambda1)*np.sin(w1*x[i]+phi1)
    y_pred[i] += alfa2*np.exp(x[i]*lambda2)*np.sin(w2*x[i]+phi2)

plt.axvline(x=idx_test, linestyle='--', color='black')
plt.plot(x_test, y_test, 'or')
plt.plot(x_train, y_train, 'og')
plt.plot(x, y_pred, 'black', lw=3, label='Mejor ajuste')
plt.ylim([100,650])
plt.title(f'Polinomio + oscilatorio + adaptación creciente (error: {np.sum(np.abs(y[idx_test:] - y_pred[idx_test:])/(144-idx_test)):.1f})')
plt.xlabel('Tiempo [Mes]')
plt.ylabel('Pasajeros')
plt.grid(False)
plt.legend();
```

## 2.2) Optimización de mínimos cuadrados

Scipy es capaz de resolver problemas no lineales de mínimos cuadrados sujetos a restricciones de la forma

\begin{align}
\text{min } \frac{1}{2} \sum_{1=1}^{m} \rho (f_i(x)^2)\\
\text{sujeto a } l_b \leq x \leq u_b
\end{align}
donde $f_i$ son funciones suaves de $R^N$ a $R$. El proposito de $\rho$ es reducir la infuencia de outliers.

Todos los métodos específicos de minimos cuadrados utilzian la matriz Jacobiana de $f$, y es muy recomendado computar esta matriz y pasarsela 'least_squares', de otra forma será estimado numericamente.

### Analisis de reacciónes enzimaticas
Se considerará el "Analysis of an Enzyme Reaction", que vamos a minimalizar con un modelo $f(x) : \mathbb{R}^{4} \to \mathbb{R}$ definido por

$$f(x_i) = x_{i,0} \frac{u^2 + x_{i,1} u}{u^2 + x_{i,2} u + x_{i,3}}$$

en que la función de costo para `scipy.optimize.least_squares` es

$$S_i = f(x_i) - y_i, \quad y_i \in \mathbb{R}$$

con $u$ el tiempo e $y$ la absorpción de las enzimas. Esta función tiene un Jacobiano definido por

$$J = \begin{bmatrix}
\frac{\partial f(x_0)}{\partial x_{0,0}} & \cdots & \frac{\partial f(x_n)}{\partial x_{n,0}} \\
\frac{\partial f(x_0)}{\partial x_{0,1}} & \cdots & \frac{\partial f(x_n)}{\partial x_{n,1}} \\
\frac{\partial f(x_0)}{\partial x_{0,2}} & \cdots & \frac{\partial f(x_n)}{\partial x_{n,2}} \\
\frac{\partial f(x_0)}{\partial x_{0,3}} & \cdots & \frac{\partial f(x_n)}{\partial x_{n,3}}
\end{bmatrix}$$

$$\begin{align}
\frac{\partial f(x_i)}{\partial x_{i,0}} &= \frac{u^2 + x_{i,1} u}{u^2 + x_{i,2} u + x_{i,3}} \\
\frac{\partial f(x_i)}{\partial x_{i,1}} &= x_{i,0} \frac{u}{u^2 + x_{i,2} u + x_{i,3}} \\
\frac{\partial f(x_i)}{\partial x_{i,2}} &= -x_{i,0} u \frac{u^2 + x_{i,1} u}{(u^2 + x_{i,2} u + x_{i,3})^2} \\
\frac{\partial f(x_i)}{\partial x_{i,3}} &= -x_{i,0} \frac{u^2 + x_{i,1} u}{(u^2 + x_{i,2} u + x_{i,3})^2} \\
\end{align}$$

```python
from scipy.optimize import least_squares

def model(x, u):
    return x[0] * (u ** 2 + x[1] * u) / (u ** 2 + x[2] * u + x[3])

def costo(x, u, y):
    return model(x, u) - y

def jac(x, u, y):
    J = np.empty((u.size, x.size))
    den = u ** 2 + x[2] * u + x[3]
    num = u ** 2 + x[1] * u
    J[:, 0] = num / den
    J[:, 1] = x[0] * u / den
    J[:, 2] = -x[0] * num * u / den ** 2
    J[:, 3] = -x[0] * num / den ** 2
    return J

u = np.array([4.0, 2.0, 1.0, 5.0e-1, 2.5e-1, 1.67e-1,
              1.25e-1, 1.0e-1, 8.33e-2, 7.14e-2, 6.25e-2])
y = np.array([1.957e-1, 1.947e-1, 1.735e-1, 1.6e-1, 8.44e-2,
              6.27e-2, 4.56e-2, 3.42e-2, 3.23e-2, 2.35e-2, 2.46e-2])
x0 = np.array([2.5, 3.9, 4.15, 3.9])
res = least_squares(costo, x0, jac=jac, bounds=(0, 100), args=(u, y), verbose=1)
```

```python
res.x
```

```python
%matplotlib inline 
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (15,5)

u_test = np.linspace(0, 5)
y_test = model(res.x, u_test)
plt.plot(u, y, 'o', markersize=4, label='observaciones')
plt.plot(u_test, y_test, label='modelo ajustado')
plt.xlabel("u")
plt.ylabel("y")
plt.title('Regresión de mínimos cuadrados para análisis de reacción enzimática')
plt.legend(loc='lower right');
```

### 2.3)  Mínimos cuadrados: repaso

El problema lineal de los minimos cuadrados aparece en muchas ramas de las matemáticas. En este problema un conjunto de coeficientes lineales se desean encontrar para poder ajustar el modelo a los datos.
En particular, se supone que los datos $y_i$ se relacionan con los datos $x_i$ por un conjunto de coeficientes $c_j$ y una función modelo $f_j(x_i)$ de la forma:

\begin{align}
y_i = \sum_{j} c_j f_j(x_i) + \epsilon_i
\end{align}

donde $\epsilon_i$ representa incertezas en los datos. La estrategia de los mínimos cuadrados es escoger coeficientes $c_j$ que minimicen el error cuadrático, es decir

\begin{align}
J(c) = \sum_{i} \left| y_i - \sum_j c_j f_j (x_i) \right| ^2
\end{align}

Definiendo $\{ A\}_{ij} = f_j (x_i)$, notemos que el problema se puede escribir $y = Ac + \epsilon$, y el problema posee solución cuando $A^H A$ es invertible y entonces

\begin{align}
c = \left( A^H A \right)^{-1} A^H y = A^{\dagger} y
\end{align}


El comando 'linalg.lstsq' resolverá el problema de los mínimos cuadrados para $c$ dado $A$ e $y$. Para calcular $A^{\dagger}$ se puede usar las funciones 'linalg.pinv' o 'linalg.pinv2'.

El siguiente ejemplo muestra el uso de las funciones antes mencionadas para el modelo

\begin{align}
y_i = c_1 e^{-x_i} + c_2 x_i
\end{align}
con $x_i=0.1$, $i = 1,...,10$, $c_1 =5, c_2 = 4$. Se añade ruido a $y_i$ y los coeficientes $c_1, c_2$ son estimados usando minimos cuadrados lineal

### 2.4) otra forma de hacer mínimos cuadrados

```python
from scipy import linalg

c1, c2 = 5.0, 2.0
i = np.r_[1:11]
xi = 0.1 * i
yi = c1 * np.exp(-xi) + c2 * xi
zi = yi + 0.05 * np.max(yi) * np.random.rand(len(yi))
print(yi)
print(zi)
```

```python
A = np.c_[np.exp(-xi)[:, np.newaxis], xi[:, np.newaxis]]
print(A)
c, resid, rank, sigma = linalg.lstsq(A, zi)

print(c, resid, rank, sigma)
print(c)
```

### 2.5) Análisis de Fourier
Con análisis de Fourier cambiamos el espacio de nuestra función desde el espacio de tiempo hasta el espacio de frecuencia. Esta transformación da la oportunidad de ver nuestros datos en una otra manera. Con dos senos y una otra signal, vemos los picos en el espacio de frecuencia ubicado en sus propias frecuencias. Por ejemplo, abajo tenemos un seno en frecuencia $20$ y uno en $32$.

```python
import scipy.fftpack

N = 600
T = 1.0 / 100.0

x = np.linspace(0, N*T, N)
y = np.sin(20 * 2.0*np.pi*x) + 0.5*np.sin(32 * 2.0*np.pi*x) + 1/(x+0.1)

xf = np.linspace(0.0, 1.0/(2.0*T), int(N/2))
yf = sp.fftpack.fft(y)


plt.plot(x, y)
plt.show()

plt.plot(xf, 2.0/N * np.abs(yf[:N//2]))
plt.show()
```
